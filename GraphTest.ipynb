```python
import matplotlib.pyplot as plt
import random
```

```python

# Generate random data for position (vertical displacement) over time
total_time = 200  # seconds
num_samples = 200  # Number of data points
max_displacement = 10.0  # meters
```

```python
# Generate random displacement data
time_points = range(0, total_time + 1, total_time // num_samples)
position_data = [random.uniform(0, max_displacement) for _ in time_points]
```

```python

# Create a Matplotlib plot
plt.figure(figsize=(10, 6))  # Optional: Adjust the figure size

# Plot the position data
plt.plot(time_points, position_data, marker='o', linestyle='-', color='b', label='Position (m)')

# Add labels and title
plt.xlabel('Time (s)')
plt.ylabel('Vertical Position (m)')
plt.title('Vertical Position vs. Time')

# Add gridlines
plt.grid(True)

# Show legend
plt.legend()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Generate time data from 0 to 10 seconds
total_time = 50  # seconds
num_samples = 200  # Number of data points
time_points = [i * (total_time / num_samples) for i in range(num_samples + 1)]

# Generate linearly increasing position data
max_position = 10.0  # meters
position_data = [i * (max_position / num_samples) for i in range(num_samples + 1)]

# Create a Matplotlib plot
plt.figure(figsize=(10, 6))  # Optional: Adjust the figure size

# Plot the position data
plt.plot(time_points, position_data, marker='o', linestyle='-', color='b', label='Position (m)')

# Add labels and title
plt.xlabel('Time (s)')
plt.ylabel('Vertical Position (m)')
plt.title('Vertical Position vs. Time')

# Add gridlines
plt.grid(True)

# Show legend
plt.legend()

# Display the plot
plt.show()
```
